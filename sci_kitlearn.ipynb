```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import metrics

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()

# Create DataFrames for features and target
x = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.DataFrame(iris.target, columns=['species'])

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, stratify=y, random_state=100)
# Creating and fitting the model
model = DecisionTreeClassifier(max_depth=3, random_state=3)
model.fit(x_train, y_train)

# Visualizing your first decision tree
plt.figure(figsize=(12, 8))  # Set the figure size
plot_tree(
    model,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True
)
plt.title('Your First Decision Tree Classifier', size=18, color='grey')
plt.show()
```

```python
print("x_train's shape is ",x_train.shape)
print("x_test's shape is ",x_test.shape)
print("y_train's shape is ",y_train.shape)
print("y_test's shape is ",y_test.shape)
```

```python
# Example: Fit a Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=100)
clf.fit(x_train, y_train)

# Predict on the test set
y_pred = clf.predict(x_test)

# Calculate and print accuracy
accuracy = metrics.accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Optional: Plot the decision tree
plt.figure(figsize=(12,8))
plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats

# Data points
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]

# Linear regression
slope, intercept, r, p, std_err = stats.linregress(x, y)

# Function to calculate the linear model values
def myfunc(x):
    return slope * x + intercept

# Creating the model data using map
mymodel = list(map(myfunc, x))

# Plotting the scatter plot and the linear regression line
plt.scatter(x, y)  # Plot the original data points
plt.plot(x, mymodel)  # Plot the linear regression line
plt.show()
```
